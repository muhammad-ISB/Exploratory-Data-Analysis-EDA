```python
# variance threshold will be removing all the low - variance features - in simple words the columns that are having 1 value in all rows or very less other values compared to a single big chunk of value
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        'a' : [1, 2, 4,3 , 5, 6],
        'b' : [1,2,3,1,2,3],
        'c' : [7,7,7,7,7,7],
        'd' : [0,0,0,0,0,0]
    }
)
df
```

```python
from sklearn.feature_selection import VarianceThreshold
var_thres = VarianceThreshold(threshold=0)
var_thres.fit(df)
```

```python
var_thres.get_support() # true true means the columns a and b are non xero variances and c,d are having 0 variance columns and constant features
```

```python
df.columns[var_thres.get_support()] # this list comprehension gives the column heading
```

```python
const_cols = [col for col in df.columns
              if col not in df.columns[var_thres.get_support()]]
const_cols # this function returns the list of columns that have constant values / they are having 0 variances
```

```python
for i in const_cols:
  print(i)
```

```python
df.drop(const_cols, axis = 1)
```
