```python
df= pd.read_csv('/content/drive/MyDrive/proj/customer.csv')
df= df.iloc[:, 2:]
df
```

```python
for i in df.columns:
  print(df[i].value_counts())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 0:2], df.iloc[:,-1], test_size = 0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [['Poor', 'Average', 'Good'],['School','UG', 'PG']]) # the ranking will be given in ascending order as in 0 is for poor and school
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```
